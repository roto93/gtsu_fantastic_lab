```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipywidgets import interact, FloatSlider
from keras.models import Sequential, Model
from keras.layers import (Input, Lambda, concatenate, Dense)
from keras.optimizers import SGD, Adam
from keras import metrics
from keras.datasets import mnist
from keras.utils import np_utils
from keras import backend as K
```

```python
(x_train0, y_train0), (x_test0, y_test0) = mnist.load_data()
x_train = x_train0.reshape(60000, -1)
x_test = x_test0.reshape(10000, -1)
```

```python
x_train -= x_train.min()
x_train = x_train/x_train.max()
```

### 我選擇用Functional API 建立模型

```python
x = Input(shape=(784, ))
enc_1 = Dense(200, activation = 'sigmoid')
enc_2 = Dense(50, activation = 'sigmoid')
enc_mean = Dense(2)
enc_var = Dense(2)
dec_1 = Dense(50, activation='sigmoid')
dec_2 = Dense(200, activation='sigmoid')
dec_out = Dense(784, activation='sigmoid')
```

```python
def sampling(args):
    m, v = args
    epsilon = K.random_normal(shape=(2,), mean=0, stddev=1)
    return m + K.exp(v / 2) * epsilon
```

```python
h1 = enc_1(x)
h2 = enc_2(h1)
h3_m = enc_mean(h2)
h3_v = enc_var(h2)
z = Lambda(sampling, output_shape=(2, ))([h3_m, h3_v])
d_h1 = dec_1(z)
d_h2 = dec_2(d_h1)
x_hat = dec_out(d_h2)
```

```python
VAE = Model(x, x_hat)
VAE.summary()
```

```python
def lossf(x,x_recon):
    loss = 784 * metrics.binary_crossentropy(x, x_recon)
    kl_loss = -0.5 * K.sum(1 + h3_v - K.square(h3_m) - K.exp(h3_v), axis= -1)
    return loss + kl_loss
```

```python
VAE.compile(loss = lossf, optimizer = Adam())
VAE.fit(x_train, x_train, batch_size = 300, epochs = 10)
```

### 訓練好了，把模型分成encoder和decoder

```python
Encoder = Model(x, h3_m)
Encoder.summary()
```

```python
h_input = Input(shape=(2, ))
Decoder = Model(h_input, dec_out(dec_2(dec_1(h_input))))
Decoder.summary()
```

### 用encoder畫圖

```python
index = np.random.randint(10000, size=1000)
latent = Encoder.predict(x_test[index])
plt.scatter(latent[:,0], latent[:,1], c=y_test0[index], cmap='tab20')
plt.colorbar()
```

### 用decoder還原資料

```python
n = 15
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
grid_x = np.linspace(-4, 4, n)
grid_y = np.linspace(-4, 3, n)

for i, yi in enumerate(grid_x):
    for j, xi in enumerate(grid_y):
        z_sample = np.array([[xi, yi]])
        x_decoded = Decoder.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[(n-i-1) * digit_size: (n - i) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit
        
plt.figure(figsize=(10, 10))
plt.imshow(figure, cmap='Greys')
plt.axis('off')
plt.show()
```

```python
idx0, idx1 = np.random.randint(60000, size=2)
def inBetween(t):
    dt0 = x_train0[idx0]
    dt1 = x_train0[idx1]
    dtt = (1-t)*x_train0[idx0] + t*x_train0[idx1]
    
    μ0 = Encoder.predict(x_train[idx0:idx0+1]).squeeze()
    μ1 = Encoder.predict(x_train[idx1:idx1+1]).squeeze()
    μt = (1-t)*μ0 + t*μ1
    
    plt.figure(figsize=(10,5))
    ax1 = plt.subplot(2,1,2)
    ax1.scatter(μ0[0], μ0[1])
    ax1.scatter(μ1[0], μ1[1])
    ax1.scatter(μt[0], μt[1])
    
    ax2 = plt.subplot(2,3,1)
    ax2.imshow(dt0, cmap='Greys')
    
    ax3 = plt.subplot(2,3,2)
    ax3.imshow(dtt, cmap='Greys')
    
    ax4 = plt.subplot(2,3,3)
    ax4.imshow(dt1, cmap='Greys')
```

```python
interact(inBetween, 
         t=FloatSlider(value=0.5, min=0, max=1.0, step=0.05,)
        )
```
