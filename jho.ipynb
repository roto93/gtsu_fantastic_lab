```python
import numpy as np   
import matplotlib.pyplot as plt   # 畫圖
import pandas as pd               # 讀表格
import locale                     # 把數字的逗點去掉 (千分位分隔符)
import matplotlib as mpl          
mpl.rc('font', family='Noto Sans CJK TC')    # 讓 matplotlib 顯示中文
%matplotlib inline
```

```python
df = pd.read_csv('../ROC_go_out.csv', encoding='big5')
```

```python
df
```

```python
locale.setlocale(locale.LC_NUMERIC,'')
```

```python
for i in range(2010,2018):
    df[str(i)]=df[str(i)].apply(locale.atof)
```

```python
df.head()
```

```python
x = np.array(df.columns[10:])
```

```python
for i in range(10):
    y = np.array(df.loc[i][10:])
    plt.plot(x,y,label=str(df.loc[i][1]))
plt.legend(loc='center', bbox_to_anchor=(1.45, 0.5),ncol=2)
plt.yticks(range(500000,5000000,500000))
plt.show()
```
